```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2

Img_path = "trxy_s2_00.tif"
Img_G_origin = cv2.imread(Img_path)
Img_number = 30
angle = 30
Img_shape = Img_G_origin.shape
rows, cols = Img_shape[:2]

for i in range(angle):
    rotation_tool = cv2.getRotationMatrix2D((cols/2, rows/2), i, 1)
    Img_G_rotated = cv2.warpAffine(Img_G_origin, rotation_tool, (cols, rows))
    cropped_Img = Img_G_rotated[80:140, 80:140]
    cv2.imwrite("Img_source\\" + str(i)+".tif", cropped_Img)

import os
import math
Img_dir = "Img_source"
file_list = sorted(os.listdir(Img_dir))
file_list.sort(key = lambda x: int(x[:-4]))

Xi = []
origin_Img_name = '0.tif'
rotation_angle = []

angle = 0
total_grey_G = []
for file_name in file_list:
    rotation_angle.append(angle)
    angle+=1
    grey_value_differences = 0.0
    total_grey_F = []
    grey_value_difference = []
    if file_name == origin_Img_name:
        rows, cols = cv2.imread(Img_dir + "\\" + file_name).shape[:2]
        grey_G = cv2.imread(Img_dir + "\\" + file_name, cv2.IMREAD_GRAYSCALE)
        for index_i in range(rows):
            for index_j in range(cols):
                total_grey_G.append(grey_G[index_i][index_j])
        total_grey_F = total_grey_G
    else:
        rows, cols = cv2.imread(Img_dir + "\\" + file_name).shape[:2]
        grey_F = cv2.imread(Img_dir + "\\" + file_name, cv2.IMREAD_GRAYSCALE)
        for index_i in range(rows):
            for index_j in range(cols):
                total_grey_F.append(grey_F[index_i][index_j])

    for i in range(len(total_grey_G)):
        grey_value_difference.append(math.pow(total_grey_G[i] - total_grey_F[i], 2))
    # print("grey_value_difference: ", grey_value_difference)
    Xi.append(np.sum(grey_value_difference))

# print(Xi)
# print(rotation_angle)
plt.plot(rotation_angle, Xi)
plt.show()
```
